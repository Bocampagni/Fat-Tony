## Função massa de probabilidade 

Função que diz a probabilidade de uma variável aleatório assumir exatamente um valor. Também conhecida como função de densidade discreta. É o primeiro recurso para descrever distribuições discretas, como a de Bernouli.

## Função densidade

É uma função que descreve a verossimilhança de uma variável aleatória tomar um valor dado. A probabilidade da variável aleatória cair em uma faixa particular é dada pela integral da densidade dessa variável sobre tal faixa. 



## Bernoulli

Distribuição discreta de espaço amostral {0,1}, ou seja, as variáveis aleatórias irão valer ou 0 ou 1. Caso o valor que está sendo testado seja 0, temos um ensaio fracassado, do contrário, um ensaio bem-sucedido.

Um exemplo clássico é o da moeda, que pode dar cara com probabilidade p, e coroa com probabilidade q = 1 - p. O experimento é justo quando p = 0.5.


```python
import random

p = 1 #Cara = 1, coroa = 0
face = random.randint(0,2)
if p == face:
    print("Ensaio bem-sucedido.")
else:
    print("Ensaio fracassou")
```

## Binomial
Se estendermos a noção de ensaios de Bernoulli para N tentativas, teremos uma distribuição discreta chamada distribuição binomial. A qual mede o número de sucessos k de N ensaios de Bernoulli com a mesma probabilidade P de ocorrência do sucesso.

```python
import random

p = 1
n = 10
binomialDist = [True if random.randint(0,2) == p else False for x in range(n)]
binomialDist
```
